## Practical Question 1 - Predict Mammal / Non-Mammal (From Scratch)

**Query:** Give Birth=yes, Can Fly=no, Live in Water=yes, Have Legs=no

```python
import numpy as np

X = [
    ['yes','no','no','yes'],['no','no','no','no'],['no','no','yes','no'],['yes','no','yes','no'],
    ['no','no','sometimes','yes'],['no','no','no','yes'],['yes','yes','no','yes'],['no','yes','no','yes'],
    ['yes','no','no','yes'],['yes','no','yes','no'],['no','no','sometimes','yes'],['no','no','sometimes','yes'],
    ['yes','no','no','yes'],['no','no','yes','no'],['no','no','sometimes','yes'],['no','no','no','yes'],
    ['no','no','no','yes'],['no','yes','no','yes'],['yes','no','yes','no'],['no','yes','no','yes']
]
y = [
    'mammals','non-mammals','non-mammals','mammals','non-mammals','non-mammals','mammals','non-mammals',
    'mammals','non-mammals','non-mammals','non-mammals','mammals','non-mammals','non-mammals','non-mammals',
    'mammals','non-mammals','mammals','non-mammals'
]
q = ['yes','no','yes','no']

classes = list(set(y))
n = len(y)
prior = {c: y.count(c)/n for c in classes}
fvals = [list(set(X[i][j] for i in range(n))) for j in range(4)]

scores = {}
for c in classes:
    Xc = [X[i] for i in range(n) if y[i]==c]
    nc = len(Xc)
    p = prior[c]
    for j in range(4):
        cnt = sum(1 for x in Xc if x[j]==q[j])
        p *= (cnt + 1) / (nc + len(fvals[j]))
    scores[c] = p

pred = max(scores, key=scores.get)
print("Naïve Bayes (From Scratch) - Animal Classification")
print("-" * 48)
print(f"Query: Give Birth=yes | Can Fly=no | Live in Water=yes | Have Legs=no")
for c in classes:
    print(f"  P({c:<14} | query) ∝ {scores[c]:.8f}")
print(f"\n  Predicted Class: {pred}")
```

## Practical Question 1 - Predict Mammal / Non-Mammal (sklearn CategoricalNB)

```python
import numpy as np
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder

X_raw = [
    ['yes','no','no','yes'],['no','no','no','no'],['no','no','yes','no'],['yes','no','yes','no'],
    ['no','no','sometimes','yes'],['no','no','no','yes'],['yes','yes','no','yes'],['no','yes','no','yes'],
    ['yes','no','no','yes'],['yes','no','yes','no'],['no','no','sometimes','yes'],['no','no','sometimes','yes'],
    ['yes','no','no','yes'],['no','no','yes','no'],['no','no','sometimes','yes'],['no','no','no','yes'],
    ['no','no','no','yes'],['no','yes','no','yes'],['yes','no','yes','no'],['no','yes','no','yes']
]
y_raw = [
    'mammals','non-mammals','non-mammals','mammals','non-mammals','non-mammals','mammals','non-mammals',
    'mammals','non-mammals','non-mammals','non-mammals','mammals','non-mammals','non-mammals','non-mammals',
    'mammals','non-mammals','mammals','non-mammals'
]
query_raw = [['yes','no','yes','no']]

encs = [LabelEncoder() for _ in range(4)]
X_enc = np.array([[encs[j].fit_transform([X_raw[i][j] for i in range(len(X_raw))])[i] for j in range(4)] for i in range(len(X_raw))])
y_enc = LabelEncoder().fit_transform(y_raw)
q_enc = np.array([[encs[j].transform([query_raw[0][j]])[0] for j in range(4)]])

model = CategoricalNB().fit(X_enc, y_enc)
pred_enc = model.predict(q_enc)[0]
pred_lbl = LabelEncoder().fit(y_raw).inverse_transform([pred_enc])[0]
proba = model.predict_proba(q_enc)[0]

print("Naïve Bayes (sklearn CategoricalNB) - Animal Classification")
print("-" * 54)
print(f"Query: Give Birth=yes | Can Fly=no | Live in Water=yes | Have Legs=no")
classes_sk = LabelEncoder().fit(y_raw).classes_
for c, p in zip(classes_sk, proba):
    print(f"  P({c:<14} | query) = {p:.6f}")
print(f"\n  Predicted Class: {pred_lbl}")
```

## Practical Question 2 - Predict Well-Behaved / Dangerous (From Scratch)

**Query:** Color=black, Fur=smooth, Size=big

```python
import numpy as np

X2 = [
    ['brown','ragged','small'],
    ['black','ragged','big'],
    ['black','smooth','big'],
    ['black','curly','small'],
    ['white','curly','small'],
    ['white','smooth','small'],
    ['red','ragged','big']
]
y2 = ['well-behaved','dangerous','dangerous','well-behaved','well-behaved','dangerous','well-behaved']
q2 = ['black','smooth','big']

classes2 = list(set(y2))
n2 = len(y2)
prior2 = {c: y2.count(c)/n2 for c in classes2}
fvals2 = [list(set(X2[i][j] for i in range(n2))) for j in range(3)]

scores2 = {}
for c in classes2:
    Xc = [X2[i] for i in range(n2) if y2[i]==c]
    nc = len(Xc)
    p = prior2[c]
    for j in range(3):
        cnt = sum(1 for x in Xc if x[j]==q2[j])
        p *= (cnt + 1) / (nc + len(fvals2[j]))
    scores2[c] = p

pred2 = max(scores2, key=scores2.get)
print("Naïve Bayes (From Scratch) - Behaviour Classification")
print("-" * 50)
print(f"Query: Color=black | Fur=smooth | Size=big")
for c in classes2:
    print(f"  P({c:<14} | query) ∝ {scores2[c]:.8f}")
print(f"\n  Predicted Class: {pred2}")
```

## Practical Question 2 - Predict Well-Behaved / Dangerous (sklearn CategoricalNB)

```python
import numpy as np
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder

X2_raw = [
    ['brown','ragged','small'],['black','ragged','big'],['black','smooth','big'],
    ['black','curly','small'],['white','curly','small'],['white','smooth','small'],['red','ragged','big']
]
y2_raw = ['well-behaved','dangerous','dangerous','well-behaved','well-behaved','dangerous','well-behaved']
q2_raw = [['black','smooth','big']]

encs2 = [LabelEncoder() for _ in range(3)]
X2_enc = np.array([[encs2[j].fit_transform([X2_raw[i][j] for i in range(len(X2_raw))])[i] for j in range(3)] for i in range(len(X2_raw))])
le2 = LabelEncoder().fit(y2_raw)
y2_enc = le2.transform(y2_raw)
q2_enc = np.array([[encs2[j].transform([q2_raw[0][j]])[0] for j in range(3)]])

model2 = CategoricalNB().fit(X2_enc, y2_enc)
pred2_enc = model2.predict(q2_enc)[0]
pred2_lbl = le2.inverse_transform([pred2_enc])[0]
proba2 = model2.predict_proba(q2_enc)[0]

print("Naïve Bayes (sklearn CategoricalNB) - Behaviour Classification")
print("-" * 58)
print(f"Query: Color=black | Fur=smooth | Size=big")
for c, p in zip(le2.classes_, proba2):
    print(f"  P({c:<14} | query) = {p:.6f}")
print(f"\n  Predicted Class: {pred2_lbl}")
```
